#### 03. If-Statements with User-Defined Functions

```python
# def price_label(row):

 # if row['prices'] <= 5: --- the colon translates to 'then'
   # return 'Low-range product' --- this is connected to the 1st condition
 # elif (row['prices'] > 5) and (row['prices'] <= 15): --- new condition, similar to the 1st condition instead of if use elif (else if)
  #  return 'Mid-range product' --- connected to the 2nd condition
 # elif row['prices'] > 15: --- less than or equal to 15 (3rd condition)
  #  return 'High range' --- return the string- links to the 3rd condition
 # else: return 'Not enough data' --- this will catch any missing values which havent been addressed anywhere in your conditions
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# lets use it

df['price_range'] = df.apply(price_label, axis=1)
```

### 06. create a new column for busiest, least busy and regularly busy

```python
# create a new column that will contain- busiest, least busy and regularly busy

result = []

for value in df_full["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). Create a new column for this using a suitable method.

4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# find out which hour of the day most orders take place

df_full['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create the result 

result_hours = []

for value in df_full["order_hour_of_day"]:
  if value in (10, 11, 14, 15, 13, 12, 16, 9):
    result_hours.append('Most orders')  
  elif value in (23, 6, 0, 1, 5, 2, 4, 3):
    result_hours.append('Fewest orders')  
  else:
    result_hours.append('Average orders')  
```

```python
# check it has added the command 

result_hours
```

```python
# create the column 

df_full['busiest_period_of_day'] = result_hours
```
